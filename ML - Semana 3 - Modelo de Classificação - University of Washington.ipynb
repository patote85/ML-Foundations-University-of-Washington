## Construindo um classificador de sentimento

```
# adicionando a coluna "word_count" para contar as palavras na coluna "review"
products['word_count'] = tc.text_analytics.count_words(products['review'])
```

```
products
```

## Definindo os positivos e negativos

```
# Primeiro iremos ignorar os reviews 3* (já que não conseguimos decidir se são positivos ou não)
products = products[products['rating']!= 3]
```

```
#sentimentos positivos ==> 4-star or 5-star reviews
# incluir uma coluna "sentiment" (binária) que para os reviews >=4 ficará com 1 e senão ficará com 0
products['sentiment'] = products['rating'] >= 4
```

```
products
```

```
# plotando a distribuição de sentimento
products['sentiment'].show()
```

## Treinando o classificador de sentimentos

```
# dividindo os dados em treinamento e teste (80 x 20)
train_data,test_data = products.random_split(.8,seed=0)
```

```
# criando o classificador sentiment_model
# usando um classificador logistico, classificando pela coluna "sentiment", usando como entrada a coluna "word_count" e para validação dos dados de teste
sentiment_model = tc.logistic_classifier.create(train_data,target='sentiment', features=['word_count'], validation_set=test_data)
```

```
#ROC curve (falso positivo, falso negativo)
sentiment_model.evaluate(test_data)
```

## Aplicando o classificador de sentimentos para entender melhor a classificação dos reviews do produto Giraffe

```
#criando a coluna "predicted_sentiment" a partir da probabilidade da predição 
products['predicted_sentiment'] = sentiment_model.predict(products, output_type = 'probability')
```

## Exercício 1

```
#Selecionar apenas algumas palavras dos reviews
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```
#função awesome_count que recebe um parametro do tipo dicionário (coluna word_count) e se a palavra "awesome" existir retorna a contagem, senão retorna 0
def awesome_count(x):
   if 'awesome' in x:
       return x['awesome']
   else:
       return 0 

#chamada da função awesome_count para cada linha de products (o método apply faz esse trabalho de "FOR") 
products['awesome'] = products['word_count'].apply(awesome_count)
```

```
#sortear de forme descendente a nova coluna "awesome" criada
products_sorted = products.sort('awesome', ascending=False)
```

```
products_sorted
```

```
#função Lambda para dado cada palavra da lista selected_words executar o .apply da função
selected_words = ['great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']

for selected in selected_words:
    products[selected] = products['word_count'].apply(lambda x: x[selected] if selected in x else 0)  
```

```
products
```
